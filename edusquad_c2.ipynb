## statistical test -> comparison between groups
แบ่งตามชุดข้อสอบที่ใช้ในการทดสอบ ได้นักเรียน 2 กลุ่ม:
- เด็กเล็ก
- เด็กกลางและเด็กโต

#### เตรียม dataframe

- df_academic_km = dataframe ที่ใช้เก็บคะแนนทักษะทางวิชาการของน้องที่อยู่ในพื้นที่โค้งรถไฟ (เด็กกลาง+โต)
- df_academic_ks = dataframe ที่ใช้เก็บคะแนนทักษะทางวิชาการของน้องที่อยู่ในพื้นที่โค้งรถไฟ (เด็กเล็ก)

```python
def add_period(df, period):
    df = df.copy()
    df["ช่วงเวลา"] = period
    return df


df_academic_km = pd.concat([
    add_period(df_academic_km_pre, "pre"),
    add_period(df_academic_km_post, "post")
])

df_academic_ks = pd.concat([
    add_period(df_academic_ks_pre, "pre"),
    add_period(df_academic_ks_post, "post")
])

df_academic_km
```

### test for normality and statistical methods

##### เด็กกลาง

```python
import pandas as pd
from scipy.stats import shapiro, ttest_rel, wilcoxon

df_mid = df_academic_km.query("ประเภทกลุ่ม == 'เด็กกลาง'")

df_wide = df_mid.pivot_table(
    index=["รหัสนักเรียน","ชื่อ","อายุ","ประเภทกลุ่ม","ทักษะ","ภาษา (คะแนนเต็ม)"],
    columns="ช่วงเวลา",
    values="คะแนน"
).reset_index()

df_wide = df_wide.dropna(subset=["pre","post"])
df_wide["diff"] = df_wide["post"] - df_wide["pre"]


results = []

for (skill, lang), subset in df_wide.groupby(["ทักษะ","ภาษา (คะแนนเต็ม)"]):
    n = len(subset)
    mean_pre = subset["pre"].mean()
    mean_post = subset["post"].mean()
    mean_diff = subset["diff"].mean()
    
    # check normality
    stat_norm, p_norm = shapiro(subset["diff"])
    
    if p_norm > 0.05: # use paired t-test
        stat, pval = ttest_rel(subset["post"], subset["pre"])
        test_used = "Paired t-test"
    else: # use wilcoxon
        stat, pval = wilcoxon(subset["post"], subset["pre"])
        test_used = "Wilcoxon Signed-Rank"
    
    results.append({
        "ทักษะ": skill,
        "ภาษา": lang,
        "n": n,
        "Pre mean": round(mean_pre,2),
        "Post mean": round(mean_post,2),
        "Mean diff": round(mean_diff,2),
        "Normality p": round(p_norm,4),
        "Test": test_used,
        "Stat": round(stat,3),
        "p-value": round(pval,4)
    })

summary_df = pd.DataFrame(results)
summary_df
```

n → จำนวนตัวอย่าง (นักเรียน) ที่มีข้อมูลทั้ง pre และ post สำหรับทักษะนั้น  
Pre mean → ค่าเฉลี่ยคะแนนก่อนเรียน (Pre-test)  
Post mean → ค่าเฉลี่ยคะแนนหลังเรียน (Post-test)  

Mean diff → ค่าความต่างเฉลี่ย = Post mean − Pre mean  
    - ถ้าเป็น บวก = โดยเฉลี่ยนักเรียนได้คะแนนเพิ่มขึ้น  
    - ถ้าเป็น ลบ = โดยเฉลี่ยนักเรียนได้คะแนนลดลง  

Stat → ค่าสถิติที่ได้จากการทดสอบ (เช่น t หรือ W)  

p-value → ค่าความน่าจะเป็นจากการทดสอบ  
    - ถ้า < 0.05 → มีนัยสำคัญทางสถิติ → แสดงว่ามีการเปลี่ยนแปลงจริง (คะแนนก่อนและหลังต่างกัน)  
    - ถ้า ≥ 0.05 → ไม่มีนัยสำคัญ → แสดงว่าคะแนนก่อนและหลังไม่ต่างกันอย่างมีนัยสำคัญ  

#### เด็กโต

```python
import pandas as pd
from scipy.stats import shapiro, ttest_rel, wilcoxon

df_old = df_academic_km.query("ประเภทกลุ่ม == 'เด็กโต'")

# pivot เพื่อเอา pre post เป็น column แทน
df_wide = df_old.pivot_table(
    index=["รหัสนักเรียน","ชื่อ","อายุ","ประเภทกลุ่ม","ทักษะ","ภาษา (คะแนนเต็ม)"],
    columns="ช่วงเวลา",
    values="คะแนน"
).reset_index()

df_wide = df_wide.dropna(subset=["pre","post"])
df_wide["diff"] = df_wide["post"] - df_wide["pre"]

results = []

for (skill, lang), subset in df_wide.groupby(["ทักษะ","ภาษา (คะแนนเต็ม)"]):
    n = len(subset)
    mean_pre = subset["pre"].mean()
    mean_post = subset["post"].mean()
    mean_diff = subset["diff"].mean()
    
    # Check normality
    stat_norm, p_norm = shapiro(subset["diff"])
    if p_norm > 0.05: # normal → paired t-test
        stat, pval = ttest_rel(subset["post"], subset["pre"])
        test_used = "Paired t-test"
    else:               # not normal → Wilcoxon
        stat, pval = wilcoxon(subset["post"], subset["pre"])
        test_used = "Wilcoxon Signed-Rank"
    
    results.append({
        "ทักษะ": skill,
        "ภาษา": lang,
        "n": n,
        "Pre mean": round(mean_pre,2),
        "Post mean": round(mean_post,2),
        "Mean diff": round(mean_diff,2),
        "Normality p": round(p_norm,4),
        "Test": test_used,
        "Stat": None if stat is None else round(stat,3),
        "p-value": None if pval is None else round(pval,4)
    })

summary_old = pd.DataFrame(results)
summary_old
```

## statistical test -> comparison between groups

##### เด็กกลาง

```python
import pandas as pd
from scipy.stats import shapiro, ttest_rel, wilcoxon

df_mid_bm = df_academic_bm.query("ประเภทกลุ่ม == 'เด็กกลาง'")

df_wide = df_mid_bm.pivot_table(
    index=["รหัสนักเรียน","ชื่อ","อายุ","ประเภทกลุ่ม","ทักษะ","ภาษา"],
    columns="ช่วงเวลา",
    values="คะแนน",
    aggfunc="first"
).reset_index()

df_wide = df_wide.dropna(subset=["pre","post"])
df_wide["diff"] = df_wide["post"] - df_wide["pre"]

results = []
for (skill, lang), subset in df_wide.groupby(["ทักษะ","ภาษา"]):
    n = len(subset)
    mean_pre = subset["pre"].mean()
    mean_post = subset["post"].mean()
    mean_diff = subset["diff"].mean()
    
    if subset["diff"].eq(0).all():  # ถ้า pre=post ทุกคน
        test_used = "No change"
        stat, pval = None, None
    else:
        stat_norm, p_norm = shapiro(subset["diff"])
        if p_norm > 0.05:
            stat, pval = ttest_rel(subset["post"], subset["pre"])
            test_used = "Paired t-test"
        else:
            stat, pval = wilcoxon(subset["post"], subset["pre"])
            test_used = "Wilcoxon Signed-Rank"
    
    results.append({
        "ทักษะ": skill,
        "ภาษา": lang,
        "n": n,
        "Pre mean": round(mean_pre,2),
        "Post mean": round(mean_post,2),
        "Mean diff": round(mean_diff,2),
        "Test": test_used,
        "Stat": None if stat is None else round(stat,3),
        "p-value": None if pval is None else round(pval,4)
    })

summary_df = pd.DataFrame(results)
summary_df
```
